Import relevant libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter 
```

Load datasets

```python
eco = pd.read_csv('economy.csv')
business = pd.read_csv('business.csv')

eco.head(), business.head()
```

```python
#adding new feature for category
eco['category'] = 'economy'
eco.shape
```

```python
business['category'] = 'business'
business.head()
```

```python
#drop head from business and merge both dataframes
df = pd.concat([eco, business], ignore_index=True)
df.head()
```

Explore Dataset

```python
df.isnull().sum()
```

```python
#explore dates
df['date'].value_counts()

#dates ranges from feb 2022 till mar 2022

#we can split and convert in month and day
df['day'] = df['date'].str.split('-').str[0]
df['month'] = df['date'].str.split('-').str[1]

df.drop('date', axis=1, inplace=True)

df.head()
```

```python
#check unique months
df['day'].unique()
```

```python
#num_code we can drop
df.drop('num_code', axis=1, inplace=True)
```

```python
#lets explore airline
df['airline'].value_counts()

#no fe needed we can directly do lable encoding on airline
```

```python
#lets explore dep_time
df['dep_time'].value_counts()

#we can solit in hour and min for departure

df['dep_hr'] = df['dep_time'].str.split(':').str[0]
df['dep_min'] = df['dep_time'].str.split(':').str[1]

#now we can drop dep_time
df.drop('dep_time', axis=1, inplace=True)
```

```python
df.head()
```

```python
Counter(df['time_taken'])
```

```python
#let's treat time taken 
df['time_taken'].sort_values()
#we can again split in hr and min
df['duration_hr'] = df['time_taken'].str.split('h ').str[0]
df['duration_min'] = df['time_taken'].str.split(" ").str[1]
df['duration_min'] = df['duration_min'].str.split("m").str[0]

#no we can drop time taken
df.drop('time_taken', axis =1, inplace = True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
#also, we can calcualte total duration in minutes

#but first convert into intergers
df['duration_hr'] = (df['duration_hr'].astype(float)).astype(int)
df['duration_min'] = df['duration_min'].replace('', 0)
df['duration_min'] = df['duration_min'].astype(int)

df['total_duration'] = (df['duration_hr']*60) + df['duration_min']
df['total_duration']
```

```python
#we can drop arrivan time and departure time and duration can be used to derive this along with duration hr and min\
df.drop(['arr_time', 'duration_hr', 'duration_min'], axis = 1, inplace = True)
```

```python
df.head()
```

```python
#lets look into ch_code
df['ch_code'].unique()
```

```python
#lets check stop feature
df['stop'].value_counts()

#let's get rid of \, n, \t first
business['stop'] = business['stop'].astype(str).str.strip()
df['stop'].value_counts()
```

```python
#let's extract stops
df['total_stop'] = df['stop'].str.strip(' ').str[0]
df['total_stop'].value_counts()

#replace n with 0 and convert in int
df['total_stop'] = df['total_stop'].replace('n', 0)
df.isnull().sum()

```

```python
#lets drop stop
df.drop('stop', axis = 1, inplace = True)
```

```python
#lets explore from and to
df['from'].value_counts(), df['to'].value_counts()
```

```python
df.info()
```

```python
#change datatypes of obj to int 
df[['day','month', 'dep_hr', 'dep_min', 'total_duration', 'total_stop']] = df[['day','month', 'dep_hr', 'dep_min', 'total_duration', 'total_stop']].astype(int)
```

```python
#let's fix price
df['price'] = df['price'].replace(',','', regex= True)

df['price'] = df['price'].astype(int)
```

```python
df.info()
```

Data is now cleaned

```python
#from str to obj
df[['airline', 'ch_code', 'from', 'to', 'category']] = df[['airline', 'ch_code', 'from', 'to', 'category']].astype(object)
```
